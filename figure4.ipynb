````python
seqs = []
for sequence in df["sequence"]:
    seqs.append(Seq.Seq(sequence))
````

````python
df["bio_seqs"] = seqs
df["protein_seqs"] = [x.translate() for x in seqs]
df
````

````python
hexons_list = []
fibers_list = []

for i, row in df.iterrows():
    curr_r = SeqRecord(
        row["protein_seqs"],
        id=row["seqnames"],
        name=row["species"] + "_" + row["product"],
        description=row["species"] + "_" + row["product"])
    
    if row["product"] == "hexon":
        hexons_list.append(curr_r)
    else:
        fibers_list.append(curr_r)
````

````python
with open("hexons.fasta", "w") as out:
    for seq in hexons_list:
        SeqIO.write(seq, out, "fasta")

# omit celo
with open("fibers.adj.fasta", "w") as out:
    for seq in fibers_list:
        if not "celo" in seq.description:
            SeqIO.write(seq, out, "fasta")
````

### aligned the sequences with muscle
```bash
/home/conradin/applications/muscle5.1 -align hexons.fasta -output hexons.afasta

/home/conradin/applications/muscle5.1 -align fibers.adj.fasta -output fibers.adj.afasta
```

````python
for protein in ["hexons", "fibers.adj"]:
    # read the aligned sequences from muslce and create a dumb consensus
    aligned_hexons = AlignIO.read("{}.afasta".format(protein), "fasta")
    summary_align = AlignInfo.SummaryInfo(aligned_hexons)
    consensus = summary_align.dumb_consensus(threshold = 0.3)
    
    # read the actual sequences from the aligned fasta file
    aligned_sequences = {}
    for hexon in SeqIO.parse("{}.afasta".format(protein), "fasta"):
        aligned_sequences[hexon.description] = hexon.seq
        
    # iterate through the consensus sequence and record the differences to the MSA
    mut_freq = np.zeros(len(consensus))

    for i, letter in enumerate(consensus):
        n = 0
        for virus in aligned_sequences:
            sequence = aligned_sequences[virus]
            if letter != sequence[i]:
                n += 1
        n = n/len(aligned_sequences) # get the mutation frequency
        mut_freq[i] = n
        
    with open("mut_freq_.NEW.{}.txt".format(protein), "w") as out:
        for i in mut_freq:
            out.write(str(i) + "\n")
````
